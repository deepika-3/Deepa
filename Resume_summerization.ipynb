```python
!pip install python-docx
import PyPDF2
import re
import docx
import spacy

nlp = spacy.load("en_core_web_sm")

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfFileReader(file)
        text = ""
        for page_num in range(reader.numPages):
            page = reader.getPage(page_num)
            text += page.extract_text()
    return text

def extract_text_from_txt(txt_path):
    with open(txt_path, 'r') as file:
        text = file.read()
    return text

def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    text = "\n".join([paragraph.text for paragraph in doc.paragraphs])
    return text

def extract_basic_info(text):
    name = extract_name(text)
    email = extract_email(text)
    phone = extract_phone(text)
    return name, email, phone

def extract_name(text):
    # This is a simplified approach, you can improve it based on your requirements
    name_pattern = re.compile(r'Name: (.+)')
    match = name_pattern.search(text)
    if match:
        return match.group(1)
    return None

def extract_email(text):
    email_pattern = re.compile(r'[\w\.-]+@[\w\.-]+')
    match = email_pattern.search(text)
    if match:
        return match.group(0)
    return None

def extract_phone(text):
    phone_pattern = re.compile(r'\b\d{10}\b|\b\d{3}[-.\s]\d{3}[-.\s]\d{4}\b')
    match = phone_pattern.search(text)
    if match:
        return match.group(0)
    return None

def extract_skills(text):
    skills = ["Python", "Java", "C++", "SQL", "Machine Learning", "Data Analysis"]  # Add more skills as needed
    extracted_skills = [skill for skill in skills if skill.lower() in text.lower()]
    return extracted_skills

def extract_experience(text):
    doc = nlp(text)
    experiences = []
    for ent in doc.ents:
        if ent.label_ in ("ORG", "WORK_OF_ART", "PRODUCT", "EVENT"):
            experiences.append(ent.text)
    return experiences

def extract_education(text):
    doc = nlp(text)
    education = []
    for ent in doc.ents:
        if ent.label_ in ("ORG", "DATE"):
            education.append(ent.text)
    return education

def summarize_resume(name, email, phone, skills, experience, education):
    summary = f"""
    Name: {name}
    Email: {email}
    Phone: {phone}

    Skills:
    {', '.join(skills)}

    Experience:
    {', '.join(experience)}

    Education:
    {', '.join(education)}
    """
    return summary

def save_summary_to_file(summary, output_path):
    with open(output_path, 'w') as file:
        file.write(summary)

def main(input_path, output_path):
    if input_path.endswith('.pdf'):
        text = extract_text_from_pdf(input_path)
    elif input_path.endswith('.txt'):
        text = extract_text_from_txt(input_path)
    elif input_path.endswith('.docx'):
        text = extract_text_from_docx(input_path)
    else:
        raise ValueError("Unsupported file format")

    name, email, phone = extract_basic_info(text)
    skills = extract_skills(text)
    experience = extract_experience(text)
    education = extract_education(text)

    summary = summarize_resume(name, email, phone, skills, experience, education)
    save_summary_to_file(summary, output_path)
    print("Summary saved to", output_path)

if __name__ == "__main__":
    input_path = "/content/S DEEPIKA New (1) (AutoRecovered).docx"  # Change to your input path
    output_path = "resume_summary.txt"  # Change to your output path
    main(input_path, output_path)
```
