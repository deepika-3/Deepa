```python
from google.colab.patches import cv2_imshow
```

```python
import cv2



cap = cv2.VideoCapture("/content/test_video.mp4")

while(cap.isOpened()):
  ret,frame = cap.read()
  frame = cv2.resize(frame,(1200,700))
  cv2_imshow(frame)
  if cv2_waitkey(10) & 0xFF == ord("q"):

    break

cap.release()
cv2.destroyAllWindows()
```
